```
# check the data
import pandas as pd
df = pd.read_csv('arrhythmia.data', header=None)
```

```
#since the data has many columns, take just the first few and name them(as per documentation)
data = df[[0,1,2,3,4,5]]
data.coloumns = ['age','sex','height','weight','QRS duration','P-R interval']
```
